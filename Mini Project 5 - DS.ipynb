## EDA

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")
```

```python
df['grad_cohort'] = pd.cut(
    df['gradyear'],
    bins=[2005, 2006, 2007, 2009],
    labels=['Senior','Middle','Junior']
)
```

```python
df['friends_category'] = pd.cut(
    df['NumberOffriends'],
    bins=[-1,49,200,df['NumberOffriends'].max()],
    labels=['Low','Medium','High']
)
```

```python
sports_cols = [
    'basketball','football','soccer','softball',
    'volleyball','swimming','baseball','tennis','sports'
]

df['sports_interest_score'] = df[sports_cols].sum(axis=1)
```

```python
arts_cols = ['dance','band','marching','music','rock']
df['arts_interest_score'] = df[arts_cols].sum(axis=1)
```

```python
appearance_cols = ['hair','dress','blonde']
df['appearance_interest_score'] = df[appearance_cols].sum(axis=1)
```

```python
def dominant_interest(row):
    scores = {
        'Sports': row['sports_interest_score'],
        'Arts': row['arts_interest_score'],
        'Fashion': row['appearance_interest_score']
    }
    return max(scores, key=scores.get)

df['dominant_interest'] = df.apply(dominant_interest, axis=1)
```

```python
df['is_sporty'] = (df['sports_interest_score'] > 0).astype(int)
```

```python
df['is_artistic'] = (df['arts_interest_score'] > 0).astype(int)
```

```python
df['is_fashion'] = (df['appearance_interest_score'] > 0).astype(int)
```

```python
def student_persona(row):
    if row['is_sporty'] and row['NumberOffriends'] > 100:
        return 'Popular Athlete'
    elif row['is_artistic']:
        return 'Creative Student'
    elif row['NumberOffriends'] < 30:
        return 'Low Social'
    else:
        return 'General Student'

df['student_persona'] = df.apply(student_persona, axis=1)
```

```python
#mencari outlier dengan metode IQR
Q1 = df[numbers].quantile(0.25)
Q3 = df[numbers].quantile(0.75)
IQR = Q3 -Q1

# Tentukan batas bawah dan atas
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

outliers = ((df[numbers] < (Q1 - 1.5 * IQR)) | (df[numbers] > (Q3 + 1.5 * IQR)))

print("Data sebelum: ", df.shape[0], "baris")
```

## Pra Modeling

### Menentukan PCA & Visualisasi Scree Plot

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=2)   # Mengambil 2 komponen utama
pca.fit(X_scaled)
X_pca = pca.transform(X_scaled)
```

```python
pcs = pca.transform(X_scaled)

df_pca = pd.DataFrame(data = pcs, columns = ['PC 1', 'PC 2'])
df_pca.head()
```

```python
pca.explained_variance_ratio_
```

```python
import numpy as np
from sklearn.decomposition import PCA

sns.set_theme(style="white")  # clean look

pca_full = PCA()
pca_full.fit(X_scaled)

expl_var = pca_full.explained_variance_ratio_
cum_var = np.cumsum(expl_var)

# cari titik elbow sederhana: perubahan (drop) terbesar
elbow_idx = np.argmax(np.diff(expl_var)) + 1  # +1 karena diff geser index

plt.figure(figsize=(12, 5))

# line plot
plt.plot(range(1, len(expl_var) + 1), expl_var, marker='o', linewidth=2)
plt.scatter(elbow_idx, expl_var[elbow_idx - 1], s=80)

# garis bantu elbow
plt.axvline(elbow_idx, linestyle='--', linewidth=1)
plt.text(elbow_idx + 0.2, expl_var[elbow_idx - 1], f"Elbow: PC{elbow_idx}", va='bottom')

plt.title("Scree Plot (Explained Variance per Principal Component)", fontsize=14, pad=12)
plt.xlabel("Principal Component")
plt.ylabel("Explained Variance Ratio")

plt.xticks(range(1, len(expl_var) + 1))
plt.grid(False)
plt.tight_layout()
plt.show()
```

## Tuning Hyperparameter

```python
# List untuk menyimpan hasil
from sklearn.metrics import silhouette_score
from sklearn.cluster import KMeans

k_values = []
silhouette_values = []

for k in range(2, 11):
    kmeans = KMeans(n_clusters=k, n_init=10, random_state=42)
    labels = kmeans.fit_predict(X_scaled)
    score = silhouette_score(X_scaled, labels)
    
    k_values.append(k)
    silhouette_values.append(score)

# Membuat tabel
df_scores = pd.DataFrame({
    "k": k_values,
    "Silhouette Score": silhouette_values
})

df_scores
```
